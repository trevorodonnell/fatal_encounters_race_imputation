### Incidents DataFrame

Checking the first ten rows of the incidents dataframe, which we will also be refered to as the main dataframe.

Here we define a function which will return the list of null values in a given dataset, sorted in the descecnding value.

```python
def null_sort(df):
     return df.apply(pd.isnull).sum().sort_values(ascending=False).loc[lambda x: x>0]
```

### Inital null value check

```python
main_nulls = null_sort(df_main)
main_nulls = pd.DataFrame(main_nulls)
main_nulls
```

### Dropping all columns with more than 1,200 null rows

```python
main_nulls_1 = main_nulls[main_nulls[0] > 1200]
main_nulls_1 = list(main_nulls_1.T.columns)
df_main = df_main.drop(columns = main_nulls_1)
```

### Dropping other unnecessary columns

```python
df_main = df_main.drop(columns = [
                        'Location of injury (address)',
                        'Location of death (city)',
                        'Location of death (zip code)',
                        'Full Address', 
                        'Latitude',
                        'Longitude',
                        'Supporting document link',
                        'Unique identifier (redundant)'])
```

### Dropping last row, which does not have relevant information

```python
df_main = df_main.drop(index = [30798])
```

### Checking for nulls again

```python
main_nulls_2 = null_sort(df_main)
main_nulls_2 = pd.DataFrame(main_nulls_2)
main_nulls_2
```

### Examining data types

```python
df_main.info()
```

### Converting date information to date-time data type

```python
df_main['Date'] = pd.to_datetime(df_main[' Date of injury resulting in death (month/day/year)'])
```

### Extracting various date intervals from date-time column

```python
df_main['year'] = pd.DatetimeIndex(df_main['Date']).year
df_main['month'] = pd.DatetimeIndex(df_main['Date']).month
df_main['week_of_year'] = df_main['Date'].dt.strftime('%U')
df_main['day_of_month'] = pd.DatetimeIndex(df_main['Date']).day
df_main['day_of_week'] = pd.DatetimeIndex(df_main['Date']).dayofweek
df_main['day_of_year'] = pd.DatetimeIndex(df_main['Date']).dayofyear
```

### Lowering the case of, as well as filling nulls of the "Agency or agencies involved" column

```python
df_main["Agency or agencies involved"] = df_main["Agency or agencies involved"].str.lower()
```

```python
df_main['Agency or agencies involved'].fillna('None', inplace= True)
```

```python
### Defining a function that will identify which type of agency was involved
```

```python
def agencyifier(x):
    x1 = str(x)
    if ',' in x1:
        return 'multiple_agencies'
    elif 'police' in x1:
        return 'police'
    elif 'sheriff' in x1:
        return 'sheriff'
    elif x1 == 'None':
        return 'None'
    else: 
        return 'other'
```

### Apply a lambda function to agency column, simplifying the type of agency

```python
df_main['Agency'] = df_main['Agency or agencies involved'].apply(lambda x: agencyifier(x))
```

### Examining distribution of agencies involved

```python
df_main['Agency'].value_counts(normalize=True)
```

### Dropping rows where the Age column is null

```python
df_main.dropna(subset = ['Age'], inplace=True )
```

### Checking for nulls again

```python
main_nulls_3 = null_sort(df_main)
main_nulls_3 = pd.DataFrame(main_nulls_3)
main_nulls_3
```

### Dropping all remainig rows with null values in the dataframe

```python
df_main.dropna(inplace=True)
```

### Final null value check for incidents dataframe

```python
main_nulls_3 = null_sort(df_main)
main_nulls_3 = pd.DataFrame(main_nulls_3)
main_nulls_3
```

## Census DataFrame

Looking at the first five rows of the census dataframe.

By examining the column names, we can see that we cover a variety of standard census topics. However, they are very patchy and inconsistent in their corresponding years.

```python
df_census.head()
```

### Examining shape of census dataframe

```python
df_census.shape
```

### Initial null value check of the census dataframe

```python
census_nulls = null_sort(df_census)
census_nulls = pd.DataFrame(census_nulls)
census_nulls
```

### Dropping all columns that have null values in more than 70 rows

```python
census_nulls1 = census_nulls[census_nulls[0] > 70]
census_nulls1 = list(census_nulls1.T.columns)
df_census = df_census.drop(columns = census_nulls1)
```

### Checking remaining nulls in census dataframe

```python
census_nulls = null_sort(df_census)
census_nulls = pd.DataFrame(census_nulls)
census_nulls
```

### Splitting dataframe columns into numeric and categorical categories

#### Isolating numerical columns by dropping categoricals.

```python
df_census_num = df_census.drop(columns = ['fips', 'state','metro_2013', 'smoking_ban_2010', 'uninsured_age_under_6_2017'])
```

### Filling numeric null values with the mean of the respective column

```python
column_means = df_census_num.mean()
df_census_num = df_census_num.fillna(column_means)
```

### Checking for nulls in the numeric subset of the 

```python
df_census_num.isnull().sum().sum()
```

#### Isolating categorical columns by dropping numerics.

```python
df_census_cat = df_census[['metro_2013', 'smoking_ban_2010']]
```

### Filling categorical nulls with 'Unknown' string

```python
df_census_cat = df_census_cat.fillna('Unknown')
```

### Dummifiying categorical variables

```python
df_census_cat = pd.get_dummies(df_census_cat, drop_first=True)
```

### Examining categorical variable dataframe

```python
df_census_cat.head()
```

### Checking categorical variable dataframe for null values

```python
df_census_cat.isnull().sum().sum()
```

### Concatenating categorical and numerical dataframes

```python
df_census1 = pd.concat([df_census_num,df_census_cat], axis = 1)
```

### Reattaching state column to new dataframe

```python
df_census1['state'] = df_census['state']
```

### Checking for null values

```python
df_census1.isnull().sum().sum()
```

### Exporting cleaned census data as CSV file

```python
df_census1.to_csv('./cleaned_data/counties2.csv')
```
